
## Objective

Evaluate models using both classification and ranking metrics.

---

## Metrics Used

- ROC-AUC (overall discrimination ability)
- F1 Score (balanced performance)
- Precision@Top15%
- Recall@Top15%

---

## Why Ranking Metrics Matter

In churn prediction:

The goal is often to target a subset of customers.

Precision@Top15% answers:
> Among top predicted churners, how many actually churn?

Recall@Top15% answers:
> How much total churn risk is captured?

These metrics directly support business decision-making.
