# 3. Time-Based Churn Definition (Leakage-Free)

## Objective

Define churn using a forward-looking, time-aware framework to simulate real-world deployment.

---

## Methodology

Cutoff Date: 2011-09-01  
Prediction Window: 90 days  

A customer is labeled as churned if:

> They make no purchase between 2011-09-01 and 2011-11-30.

---

## Why Time-Based Labeling?

Traditional churn modeling often uses recency directly in the target definition, causing target leakage.

This implementation:
- Separates feature window (historical data)
- From prediction window (future data)
- Ensures no future information leaks into features
