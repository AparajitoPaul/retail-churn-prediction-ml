# 2. Data Cleaning & Transaction Preparation

## Objective

This notebook prepares raw transaction data for modeling.

Cleaning ensures:
- Invalid transactions are removed
- Revenue values are computed correctly
- Only identifiable customers are retained

---

```python
# Import Setup
import sys
import os
sys.path.append(os.path.abspath(".."))

import pandas as pd
from src.config import RAW_DATA_PATH, PROCESSED_DATA_PATH
from src.preprocessing import clean_data
```

```python
# Loading Raw Data
df = pd.read_csv(RAW_DATA_PATH, encoding="ISO-8859-1")
```

## Cleaning Steps

1. Remove cancelled invoices.
2. Remove negative quantities and prices.
3. Drop rows with missing Customer ID.
4. Convert InvoiceDate to datetime format.
5. Create Revenue column:
   
   Revenue = Quantity × UnitPrice

---

```python
# Cleaning Data
df_clean = clean_data(df)

print("Original shape:", df.shape)
print("Cleaned shape:", df_clean.shape)

df_clean.head()
```

```python
output_path = os.path.join(PROCESSED_DATA_PATH, "cleaned_retail.csv")
df_clean.to_csv(output_path, index=False)

print("Cleaned data saved successfully.")
```

- Cancelled transactions
